```python
# In this function,the number of components or the rank, and the image itself are passed in as arguments. The original image is converted
# into an numpy array, reshaed into 2 dimensions, pca is applied, then a model is fit with the reshaped 2D array and dimensionaility reduction is
# applied. Then the data is transformed back to its original space. Finally, the array is reshaped back to the original 3 dimensional 
# image with dimensionality reduction applied. The reshaped image with the dimensionality reduction applied is returned. 
# Assume the input images has dimension of (1080,1920,3)

def pca_image(n_components,original_img):
    # convert the image into an numpy array
    image_array = np.array(original_img,dtype=np.float64)/255
    # reshape the image into 2 dimensions so PCA can understand
    original_array_reshape = np.reshape(image_array, (1080, 3*1920))
    
    # Applying the PCA with the number of components passed in the pca function and the solver being set to auto 
    pca = PCA(n_components, svd_solver='auto')
    # fit the model with the reshaped 2 dimernsion array and apply dimensionality reduction
    img_pca = pca.fit_transform(original_array_reshape)
    
    # Transform the data back to its original space
    img_restored = pca.inverse_transform(img_pca)
    
    # reshape back to the original 3 dimensional image with dimensionality reduction applied
    reshaped_img = np.reshape(img_restored, (1080,1920,3))
    
    # The reshaped image with the dimensionality reduction applied will be returned 
    return reshaped_img
     
```

```python
# driver code, image 1, rank = 10
X = pca_image(10, img)
plt.axis('off')
plt.imshow(X)
```
